 **Gradient Descent Algorith**:
 It is a algorithm to find out the best fitting line for given trainning data set.

 it is a optimal technique to minimize the cost function of a model.

```python
# prompt: give types of gradient descent algorithms

#  **Gradient Descent Algorithms**:
#  These are algorithms to find the best fitting line (or hyperplane) for a given training dataset.
#  They are optimization techniques to minimize the cost function of a model.

# Types of Gradient Descent Algorithms:

# 1. Batch Gradient Descent:
#    - Computes the gradient of the cost function using the entire training dataset in each iteration.
#    - Guarantees convergence to a local minimum for convex functions and to a local minimum for non-convex functions.
#    - Can be slow for very large datasets.

# 2. Stochastic Gradient Descent (SGD):
#    - Computes the gradient of the cost function using a single training example in each iteration.
#    - Faster than Batch Gradient Descent for large datasets due to frequent updates.
#    - More noisy updates, leading to oscillations around the minimum.
#    - Can escape shallow local minima.

# 3. Mini-Batch Gradient Descent:
#    - Computes the gradient of the cost function using a small subset (mini-batch) of the training dataset in each iteration.
#    - Balances the speed of SGD with the stability of Batch Gradient Descent.
#    - The most commonly used variant in practice.

# Other variations and improvements to these basic algorithms exist, including:
# - Momentum: Helps accelerate SGD in relevant directions and dampens oscillations.
# - Nesterov Accelerated Gradient (NAG): An enhanced version of momentum.
# - Adagrad: Adapts the learning rate for each parameter based on historical gradients.
# - Adadelta: An extension of Adagrad that addresses its monotonically decreasing learning rate.
# - RMSprop: Another extension of Adagrad that improves its performance.
# - Adam: Combines ideas from momentum and RMSprop.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 50  # Number of iterations
    n = len(x)
    learning_rate = 0.08

    # Plot original data points
    plt.scatter(x, y, color='blue', label="Data Points")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Gradient Descent: Regression Lines Over Iterations")

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        #cost function
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])
        #formula
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

        # Plot regression line for the current iteration
        if i%10==0:
          plt.plot(x, y_predicted, label=f"Iteration {i}, Cost: {cost:.2f}")

    # Add legend to display iteration info
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

    print(f"Final parameters: m = {m_curr}, b = {b_curr}")

# Example dataset
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

gradient_descent(x, y)


# To keep the last graph displayed
plt.show()
```
